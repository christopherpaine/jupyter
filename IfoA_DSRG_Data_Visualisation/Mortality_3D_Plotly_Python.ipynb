```python
#useful article on different ways to load external data into colab:
#https://towardsdatascience.com/7-ways-to-load-external-data-into-google-colab-7ba73e7d5fc7

import plotly.offline as pyo
import plotly.graph_objs as go
import pandas as pd
from scipy.interpolate import griddata
import numpy as np

df = pd.read_csv('https://raw.githubusercontent.com/christopherpaine/MDD/main/Mortality_tables/HMD_UK_males_1x1.csv',header=0)
x, y, z = (np.array(df[df['Age']!='110+'][col], dtype=float) for col in ['Age', 'Year', 'qx'])
xi, yi = np.linspace(x.min(), x.max(), 100), np.linspace(y.min(), y.max(), 100)
X, Y = np.meshgrid(xi, yi)
Z = griddata((x,y),z,(X,Y), method='cubic')
fig = go.Figure(go.Surface(x=xi,y=yi,z=Z,colorscale ='Blues'))
fig.show()


```

```python
#now using log of mortality rate
z = np.log(z)
Z = griddata((x,y),z,(X,Y), method='cubic')
fig = go.Figure(go.Surface(x=xi,y=yi,z=Z,colorscale ='Blues'))
fig.show()
```
